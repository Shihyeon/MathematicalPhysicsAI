```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from torch import nn
```

$
d\omega = 0.1
$

$
\phi = 0.01
$

$
\Omega \times V_{0} = 1
$

$
T = 100000
$

$
\Delta t = 0.001
$

$
U(t=0) = 
\begin{pmatrix}
1 & 0 \\
0 & 1 \\
\end{pmatrix}
$

```python
dw = 0
phi = 0
om_v0 = 1
dt = 0.001

u0 = torch.tensor([[1, 0], [0, 1]]) * (1. + 0j)
```

$
H(t) = -\frac{1}{2} \Omega V_{0} s(t) 
\begin{pmatrix}
0                        & e^{i(\text{d} \omega t + \phi)} \\
e^{-i(\text{d} \omega t + \phi)} & 0                       \\
\end{pmatrix}
$

```python
def H(s_t, T):
    t = T*dt
    y = -0.5 * om_v0 * s_t * torch.tensor([[0, torch.exp(torch.tensor(+1j * (dw*t + phi)))], 
                                           [torch.exp(torch.tensor(-1j * (dw*t + phi))), 0]])
    return y
```

```python
a = 1
class NeuralNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fcn = nn.Sequential(nn.Linear(a*8, a*16),
                                 nn.ReLU(),
                                 nn.Linear(a*16, a*16),
                                 nn.ReLU(),
                                 nn.Linear(a*16, a*8),)  # ReLU가 무엇이지??
    
    def forward(self, x):
        y = self.fcn(x)
        return y
```

```python
model = NeuralNet()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)  # lr = learning rate
```

```python
# pred = model(torch.randn(1024*8))
# print(pred.shape)

# torch.Size([8192])
```

```python
loss_list = []

for epoch in range(1, 1000+1):
    input_t = torch.arange(a*8)*dt
    model.train()
    st = model(input_t)
    ut = u0.detach().clone()

    for T in range(a*8):
        dudt = -1j*torch.matmul(H(st[T], T), ut)
        ut += dudt*dt
    
    p = torch.square(torch.matmul())  # p 정의하기

    loss = -torch.log(p).requires_grad_(True)
    loss_list.append(loss.item())
    # loss = torch.sum(st**2).requires_grad_(True)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if epoch % 5 == 1:
        
        print(loss)
        print(ut)
        plt.plot(st.detach().numpy(), 'b.')
        plt.show()
```
